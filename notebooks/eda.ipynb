```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set Seaborn style for better visuals
sns.set(style="whitegrid")

```

```python
# Load the dataset
df = pd.read_csv('../data/International_tourism.csv')

# Show the first 5 rows
df.head()
```

```python
# Shape of the dataset (rows, columns)
print("Shape of the dataset:", df.shape)

# Info about columns and datatypes
df.info()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Summary stats for numeric columns
df.describe()
```

```python
df.columns
```

```python
# Reshape from wide to long format
df_long = df.melt(
    id_vars=['Country', 'Report Type', 'Category', 'Subcategory', 'Metric'],
    var_name='Year',
    value_name='Value'
)

# Convert Year to numeric
df_long['Year'] = pd.to_numeric(df_long['Year'])

# Drop missing values if any
df_long = df_long.dropna()

# Show the reshaped data
df_long.head()
```

```python
# Filter for United Arab Emirates
uae_data = df_long[df_long['Country'] == 'United Arab Emirates']

# Plot the trend
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Value', data=uae_data, marker='o')
plt.title('International Tourism in United Arab Emirates (1995â€“2022)')
plt.xlabel('Year')
plt.ylabel('Tourism Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# See all unique country names in the dataset
df_long['Country'].unique()
```

```python
# Filter for Egypt
egypt_data = df_long[df_long['Country'] == 'EGYPT']

# Plot the trend
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Value', data=egypt_data, marker='o')
plt.title('International Tourism in Egypt (1995â€“2022)')
plt.xlabel('Year')
plt.ylabel('Tourism Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
